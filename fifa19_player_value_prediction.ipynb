## 1. Setup & Data Loading

Import libraries and load the FIFA 19 dataset.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
```

```python
df=pd.read_csv('../data/fifa19.csv')
df.head()
```

```python
le = LabelEncoder()
df['Position'] = le.fit_transform(df['Position'])

df.head()
```


## 3. Data Preprocessing

Handle missing values, encode features, and prepare for modeling.


```python
df.drop(columns=['Name','Nationality','Club'],axis=1,inplace=True)
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
plt.figure(figsize=(30,30))
corr=df.drop('Value',axis=1).corr()
sns.heatmap(corr, annot=True, cmap="Blues")
```


## 4. Model Training — SVR

Train Support Vector Regression and Linear Regression models, then compare performance.


```python
X = df.drop('Value', axis=1)
y = df['Value']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
svr= SVR(kernel='rbf')
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(mae)
print(r2)
```
