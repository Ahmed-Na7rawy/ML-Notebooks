```python
import copy
original = [1,2,[3,4]]
shallow_copy = copy.copy(original)
shallow_copy[2][0] = 99
print(original)
print(shallow_copy)
```

```python
import copy
original = [1,2,[3,4]]
deep_copy = copy.deepcopy(original)
deep_copy[2][0] = 99
print(original)
print(shallow_copy)
```
