# **matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Salary_Data.csv')
df.head()
```

```python
plt.plot(df['YearsExperience'],df['Salary'] , 'g--')
plt.xticks(rotation = 45)
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.bar(df['YearsExperience'],df['Salary'], width=.5)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5 , alpha = 0.5)
plt.show()
```

```python
plt.scatter(df['YearsExperience'],df['Salary'], color = 'red')
plt.show()
```

```python
plt.hist(df['Salary'])
plt.show()
```

```python
df.hist(figsize=(10,5))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df['YearsExperience'],df['Salary'] , 'g--')
plt.subplot(1,2,2)
plt.bar(df['YearsExperience'],df['Salary'], width=.5)
plt.show()
```

```python
cars = ['Ford', 'BMW', 'Volvo' , 'audi', 'tesla']
values = [20 , 50, 60 , 30 , 75]
colors=['brown','cyan' , 'blue' , 'indigo' , 'gray']

plt.pie(values , labels=cars , colors=colors , explode=[.1 ,.2 , .05 , .1 , .05] , autopct='%1.0f%%' , shadow=True)
plt.legend(cars , loc = 'center right' , bbox_to_anchor=(1, 0, 0.5, 1))  #bbox_to_anchor = (x, y, width, height) x=at the right y=at the bottom

plt.show()
```

```python
df = np.random.normal(loc=0 , scale=1 , size=100)
q1 , q3 = np.percentile(df , [25 , 75])
IQR = q3 - q1
lower_bound = q1 - (1.5 * IQR)
upper_bound = q3 + (1.5 * IQR)
outlier = (df < lower_bound) | (df > upper_bound)
print(lower_bound)
print(upper_bound)
print(df[outlier])

# Drop outliers — keep only values within bounds
df_no_outliers = df[~outlier]  # ~ flips the boolean mask  ~gives the inverse — the values you want to keep.

# Print cleaned data
print("Data without outliers:", df_no_outliers)
```
