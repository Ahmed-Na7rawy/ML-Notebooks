Exercise: "Even Index Encoder"

1.Take an even-length input from the user.

2.Remove odd-indexed characters (keep 0, 2, 4...).

3.Reverse the last half of the result. (abcd->abdc)

4.Swap the first and last character.

5.Print the encoded message in UPPERCASE.

example of output:

input: python

output: TOP

```python
input_string = input()
stepped_word = input_string[::2]
mid = len(stepped_word) // 2
reversed_word = stepped_word[:mid] + stepped_word[mid:][::-1]
chars = list(reversed_word)
chars[0], chars[-1] = chars[-1], chars[0]
reversed = "".join(chars)
print(reversed.upper())
```

Exercise: "List Comprehension Challenges"

1.Squares of evens: [4, 16, 36, 64, 100, 144, 196, 256, 324, 400]

```python
squares_of_evens = [x**2 for x in range(1, 21) if x % 2 == 0]
print(f"Squares of evens: {squares_of_evens}")
```

2.No vowels: "Hello World" → "Hll Wrld"
(use not in)

```python
input_string_vowels = input()
vowels = "aeiouAEIOU"
no_vowels_string = "".join([char for char in input_string_vowels if char not in vowels])
print(no_vowels_string)
```

Exercise: Shortening long names

1.Create a list called names like this: names =

[‘Ahmed’
,
‘Abdelrahman’
,
‘Mona’]

2.Filter the long names that are longer than 4 letters.

3.Map those name to their first 3 letters (a short nickname)

Input : [‘Ahmed’
,
‘Abdelrahman’
,
‘Mona’]

output: [‘Ahm’
,
‘Abd’]

```python
names = input().split(',')
nicknames = [name[:3] for name in names if len(name) > 4]
print(nicknames)
```
