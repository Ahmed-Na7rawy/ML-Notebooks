## 1. Setup & Data Loading

Import libraries and load the credit card transaction dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
from sklearn.metrics import classification_report, confusion_matrix
```

```python

df = pd.read_csv('../data/creditcard.csv')
df.head
```


## 2. Data Exploration

Explore the highly imbalanced dataset and understand feature distributions.


```python
df.describe()
```

```python
df.info()
```


## 3. Preprocessing & Feature Scaling

Standardize features and prepare data for anomaly detection.


```python
df = df.drop(['Time'], axis=1)
```

```python

scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
```


## 4. One-Class SVM Training

Train the anomaly detection model and evaluate performance.


```python
train_data = df[df['Class'] == 0].sample(frac=0.1, random_state=42).drop(['Class'], axis=1)
model = OneClassSVM(kernel='rbf', nu=0.01)
model.fit(train_data)
```

```python
predictions = model.predict(df.drop(['Class'], axis=1))
y_true = df['Class'].map({0: 1, 1: -1})

print(classification_report(y_true, predictions))
```

```python

cm = confusion_matrix(y_true, predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.show()
```
